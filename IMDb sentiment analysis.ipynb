```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
```

```python
np.random.seed(42)
hist={'loss':np.random.randint(100,size=15),'val_loss':np.random.randint(50,size=15)}
plt.plot(hist['loss'],label="General Loss")
plt.plot(hist['val_loss'],label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
from pathlib import Path
```

```python
image_path=Path('./imgs')
```

```python
for images in image_path.glob('*.png'):
    image_path_str=str(images)
    image_title=image_path_str.split("\\")[-1]
```

```python
print(image_path_str)
print(image_title)
```

```python
import cv2
```

```python
img=cv2.imread(image_path_str)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img=cv2.resize(img,(40,40))
plt.imshow(img)
plt.axis("off")
plt.title(image_title)
plt.show()
```
