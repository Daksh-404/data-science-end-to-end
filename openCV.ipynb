# SnapChat Filter


### 1. Data Preprocessing

```python
# reading the image
img=cv2.imread(r"D:\dataScience\machine-learning-online-2018-master\Datasets\snapChatFilter\Before.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#reading other images
glasses=cv2.imread(r"D:\dataScience\machine-learning-online-2018-master\Datasets\snapChatFilter\glasses.png")
mustache=cv2.imread(r"D:\dataScience\machine-learning-online-2018-master\Datasets\snapChatFilter\mustache.png")
#showing the image
plt.imshow(img)
plt.show()
```

### 2. Applying the Cascades/classifiers

```python
# intiating the cascades
faces=cv2.CascadeClassifier(r"D:\dataScience\machine-learning-online-2018-master\6. Project - Face Recognition\Face Recognition Project\haarcascade_frontalface_alt.xml")
eyes=cv2.CascadeClassifier(r"D:\dataScience\machine-learning-online-2018-master\Datasets\snapChatFilter\third-party\frontalEyes35x16.xml")
nose=cv2.CascadeClassifier(r"D:\dataScience\machine-learning-online-2018-master\Datasets\snapChatFilter\third-party\Nose18x15.xml")
# finding the coordinates for eyes and the nose and placing the bounding box
for (fx,fy,fw,fh) in faces.detectMultiScale(img,1.3,5):
    cv2.rectangle(img,(fx,fy),(fx+fw,fy+fh),(0,255,0),2)
    for (ex,ey,ew,eh) in eyes.detectMultiScale(img):
            cv2.rectangle(img,(ex,ey+22),(ex+ew,ey+eh-20),(255,255,0),2)
    for (ex,ey,ew,eh) in nose.detectMultiScale(img):
            cv2.rectangle(img,(ex,ey),(ex+ew,ey+eh),(0,255,255),2)
plt.imshow(img)
plt.show()
```
