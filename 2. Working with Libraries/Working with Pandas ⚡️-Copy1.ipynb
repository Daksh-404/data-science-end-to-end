```python
## Try to gain some insight - Plot X-Len(MovieName), Y-Freq
```

```python
titles = df.get('movie_title')
#print(titles)
```

```python
freq_title = {}

for t in titles:
    length = len(t)
    
    try:
        freq_title[length] += 1
    except:
        freq_title[length] = 1

print(freq_title)
```

```python
X  = np.array(list(freq_title.keys()))
Y = np.array(list(freq_title.values()))
```

```python
plt.scatter(X,Y)
plt.xlabel("Length Title")
plt.ylabel("No of movies")
plt.show()
```
